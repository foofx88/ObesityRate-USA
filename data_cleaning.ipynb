```python
import pandas as pd
import os
```

```python
#initialise process to merge related CSVs into one

ff_data = pd.read_csv("datasets\FF_Restaurants_Per_Capita_US-States.csv")
commute_data = pd.read_csv("datasets\commute_data.csv")
active_data = pd.read_csv("datasets\obese_vs_activeness.csv")
age_data = pd.read_csv(r"datasets\age_obesity.csv")


```

```python
#data exploration
ff_data.head(5)
```

```python
#data exploration
active_data.head(5)
```

```python
#from the 2 observations, both ff_data and active_data can be merged through alpha code

merged_one = pd.merge(active_data, ff_data, on='alpha code', how='right')

#once merge, need to drop duplicated columns and 'District of Columnbia'
merged_one = merged_one.drop(columns=['State Name', 'state code'], index=50)

#data check
merged_one.tail(5)
```

```python
#data exploration
age_data.head(5)
```

```python
#data exploration
commute_data.head(5)
```

```python
'''
further exploration of age_data and commute data needed
more transformation where age_data needs to be cleaned up
and commute_data['Mode'] needs to be made to columns so 
commute_data['State'] does not repeat itself
'''

pivot_commute = commute_data.pivot(index='State', columns='Mode', values='Commute mode share (percent)')
pivot_commute.reset_index(inplace=True)

#as we do not want to include District of Columbia and Puerto Rico
pivot_commute = pivot_commute.drop(pivot_commute.index[[8,39]])


#data check to ensure unwanted state removed
pivot_commute
```

```python
#to remove % symbols
pivot_commute=pivot_commute.replace('%','',regex=True)
```

```python
#commute data(pivot_commute) can now be merged with merged_one
merged_two = pd.merge(merged_one, pivot_commute, on='State', how='right')

#data check
merged_two.head(5)
```

```python
#As the age_data dataframe has a lot of data, needed to rearrange the columns for easier understanding
age_data.rename(columns={'Unnamed: 0':'State'}, inplace = True)

#view all column names, then rearrange according to logical order
age_data.columns
```

```python
# rearrange column headers according to categories sex, income level, age group, race, educational level
age_data = age_data[['State', 'Obesity','Male', 'Female','Less Than $25,000', '$25-$49,999', '$50-$74,999', '$75,000 or More',
       'Ages 18-44', 'Ages 45-64', 'Ages 65+', 'White', 'American Indian/Alaska Native',
       'Asian', 'Black', 'Hispanic', 'Multiracial', 'Hawaiian/Pacific Islander', 'Other Race', 
        'High School Grad','Less Than High School', 'College Grad','Some College']]
# to drop off District of Columbia
age_data = age_data.drop(index=7)
```
